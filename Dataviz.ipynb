# La visualisation des donn√©es

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Graphe 1

```python
# Configuration de Seaborn pour une meilleure apparence with '.set'
sns.set(style="whitegrid", font_scale=1)

# Cr√©er les donn√©es pour la droite d'√©quation x = 3y
y_values = np.linspace(0, 50/3, 100) #100 valeurs √©quidistances allant de 0 √† '50/3'
x_values = 3 * y_values

# Cr√©ation de la figure et d'un axe ('ax1')
fig, ax1 = plt.subplots()

# Tracer la droite
ax1.plot(x_values, y_values, color='red', label='x = 3y')

# D√©finition des libell√©s
ax1.set_xlabel('Axe des x')
ax1.set_ylabel('Axe des y')
ax1.set_title('Graphe 1')

# Ajoute la l√©gende
ax1.legend()

# Afficher le graphique
plt.show()
```

### Graphe 2

```python
sales_df = pd.read_csv("company_sales_data.csv")
display(sales_df)
```

#### Un graphe qui donne la courbe du profit total r√©alis√© chaque mois

```python
# Supposons que vous ayez un dictionnaire de correspondance des num√©ros de mois aux noms de mois
mois_dict = {
    1: 'Jan.',
    2: 'F√©v.',
    3: 'Mars',
    4: 'Avr.',
    5: 'Mai',
    6: 'Juin',
    7: 'Juil.',
    8: 'Ao√ªt',
    9: 'Sept.',
    10: 'Oct.',
    11: 'Nov.',
    12: 'D√©c.'
}

y_values = sales_df['total_profit']
x_values = list(range(1, 13))

fig, ax1 = plt.subplots()

ax1.plot(x_values, y_values, color='C4', marker='o', label='Profit total par mois')

# Remplacer les valeurs num√©riques de l'axe des abscisses par les noms des mois
ax1.set_xticks(x_values)  # Les positions des marqueurs sur l'axe des abscisses
ax1.set_xticklabels([mois_dict[num] for num in x_values], rotation=45, ha='right')  # Les noms des mois correspondants

ax1.set_xlabel('Mois')
ax1.set_ylabel('Profit total (en ‚Ç¨)')
ax1.set_title('Graphe 2')

ax1.legend()

plt.show()
```

### Graphe 3

#### Un graphe qui donne la courbe de profit pour chaque produit, en fonction du temps

```python
x = list(range(1, 13))

fig, ax = plt.subplots()

ax.plot(x, sales_df.iloc[:, 1:7]*10, label=['Cr√®me visage','Nettoyant visage','Dentifirce','Savon','Shampoing','Cr√®me de corps'], linestyle='-', marker='.' )

#facewash and moisturizer are over each other

# To convert num in month name
ax.set_xticks(x)  
ax.set_xticklabels([mois_dict[num] for num in x], rotation=45, ha='right')

ax.set_xlabel('Mois')
ax.set_ylabel('Profit total (en ‚Ç¨)')
ax.set_title('Graphe 3')

ax.legend(fontsize=7, ncol=2)
plt.show()
```

### Graphe 4

#### Un graphe qui compare, chaque mois, le nombre d'unit√©s vendus pour chaque produit. √Ä vous de choisir la meilleure repr√©sentation pour conna√Ætre le produit le plus vendu par mois.

```python
product_columns = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']

# Cr√©ation du graphique
plt.figure(figsize=(12, 6))  # D√©finir la taille du graphe

# Initialisation des positions des barres sur l'axe X
x_positions = range(len(sales_df['month_number']))

# Boucle sur chaque produit pour cr√©er les barres empil√©es
bottom = [0] * len(sales_df['month_number'])  # Variable pour garder une trace de la hauteur de la barre pr√©c√©dente
for product in product_columns:
    plt.bar(
        x_positions,
        sales_df[product],
        bottom=bottom,
        label=product
    )
    bottom = [bottom[i] + sales_df[product][i] for i in range(len(sales_df['month_number']))]

# Configurations suppl√©mentaires du graphe
plt.xlabel('Mois')
plt.ylabel('Nombre d\'unit√©s vendues')
plt.title('Comparaison des ventes mensuelles par produit')
plt.legend(loc='upper left', ncol=2)
plt.xticks(x_positions, sales_df['month_number'])  # Utilise les mois sur l'axe X
plt.tight_layout()  # Pour √©viter les chevauchements de texte

# Afficher le graphe
plt.show()
```

### Graphe 5

#### Un graphe qui donne la contribution de chaque produit dans le profit total (sur l'ann√©e √©tudi√©e) en  pourcentage pour chaque produit.

```python
ca = sales_df['total_profit'].sum()

def product_part(product):
    product = product.sum()*10/ca*100
    product_round = product.round(0).astype(int)
    return product_round

columns_to_calculate = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']

# Calcul des parties pour chaque colonne √† l'aide d'une boucle
results = {}
for column in columns_to_calculate:
    results[column] = product_part(sales_df[column])

# R√©cup√©rer les noms des colonnes et les valeurs calcul√©es √† partir de results.items()
labels, values = zip(*results.items())

# Cr√©er le camembert
plt.figure(figsize=(8, 5))  # Changer la taille du graphique si n√©cessaire
plt.pie(values, labels=labels, autopct='%1i%%', startangle=140)
plt.title('Part de chaque produit dans le CA total')
plt.axis('equal')  # Assurer un affichage circulaire
plt.show()
```
